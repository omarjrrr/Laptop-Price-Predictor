```python
def type_processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    elif text.split()[0] == 'Intel':
            return 'Other Intel Processor'
    else:
      return "AMD processor"
```

```python
df['Cpu Brand'] = df['Cpu Name'].apply(type_processor)
```

```python
df['Memory'] = df['Memory'].astype(str).replace(r'\.0$', '', regex=True)
df['Memory'] = df['Memory'].str.replace(r'\s+', '', regex=True)
df['Memory'] = df['Memory'].str.replace('GB', '', regex=False)
df['Memory'] = df['Memory'].str.replace('TB', '000', regex=False)
```

```python
new = df['Memory'].str.split('+', n=1, expand=True)
df['first'] = new[0].str.strip()
df['second'] = new[1]
df['second'].fillna('0', inplace=True)
```

```python
for layer in ['first','second']:
    df[f'Layer_{layer}_HDD'] = df[layer].apply(lambda x: 1 if 'HDD' in str(x) else 0)
    df[f'Layer_{layer}_SSD'] = df[layer].apply(lambda x: 1 if 'SSD' in str(x) else 0)
    df[f'Layer_{layer}_Hybrid'] = df[layer].apply(lambda x: 1 if 'Hybrid' in str(x) else 0)
    df[f'Layer_{layer}_Flash'] = df[layer].apply(lambda x: 1 if 'Flash Storage' in str(x) else 0)
```

```python
df['first_num'] = df['first'].str.extract(r'([\d\.]+)')
df['second_num'] = df['second'].str.extract(r'([\d\.]+)')
df['first_num'] = df['first_num'].astype(float).fillna(0)
df['second_num'] = df['second_num'].astype(float).fillna(0)
df['HDD'] = df['first_num']*df['Layer_first_HDD'] + df['second_num']*df['Layer_second_HDD']
df['SSD'] = df['first_num']*df['Layer_first_SSD'] + df['second_num']*df['Layer_second_SSD']
df['Hybrid'] = df['first_num']*df['Layer_first_Hybrid'] + df['second_num']*df['Layer_second_Hybrid']
df['Flash_Storage'] = df['first_num']*df['Layer_first_Flash'] + df['second_num']*df['Layer_second_Flash']
df[['HDD','SSD','Hybrid','Flash_Storage']] = df[['HDD','SSD','Hybrid','Flash_Storage']].fillna(0).astype(int)
```

```python
df.drop(columns=[c for c in df.columns if c.startswith('Layer_') or c in ['first','second','first_num','second_num']], inplace=True)
```

```python
x = df.drop(columns=['Price'])
y = np.log(df["Price"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=35)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0,1,7,10,11])
], remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

pipe.fit(x_train, y_train)
```

```python
y_pred =pipe.predict(x_test)
```

```python
print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
